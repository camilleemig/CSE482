```python
import pandas as pd

data = pd.read_csv('ml-latest-small/ratings.csv')
movies = pd.read_csv('ml-latest-small/movies.csv')
```

```python
users = {}
for user in set(data['userId']):
    users[user] = {}
    for index, row in data.loc[data['userId'] == user].iterrows():
        movieName = movies.loc[movies['movieId'] == row['movieId']].iloc[0]
        movieName = movieName['title']
        users[user][movieName] = row['rating']
```

```python
sorted_users = sorted(users.keys())
test_user = sorted_users[0]
test_users_movies = set(users[test_user].keys())
test_users_movies_with_ratings = users[test_user]
shared_movies = {}
for user in sorted_users[1:]:
    shared_movies[user] = test_users_movies & set(users[user].keys())
shared_movies
```

```python
from math import sqrt
def find_cosine_similarity(ratings1, ratings2):
    d1 = 0
    d2 = 0
    dot = 0
    for movie in ratings1:
        rating1 = ratings1[movie]
        rating2 = ratings2[movie]
        dot += rating1*rating2
        d1 += rating1*rating1
        d2 += rating2*rating2
    return dot / (sqrt(d1)*sqrt(d2))

similarities = {}
for user in sorted_users[1:]:
    user1_movies = dict([k, v for k,v in test_users_movies_with_ratings.items() if k in shared_users[user]])
    similarity = find_cosine_similarity(test_users_movies_with_ratings, users[user])
    similarities[user] = similarity
similarities
```
